```python
def c(x,y):
    return np.cos(np.sqrt(x**2+y**2))
x = np.linspace(5,-5,100)
y = np.linspace(5,-5,100)
X,Y = np.meshgrid(x,y)
z = c(X,Y)
A = plt.axes(projection = '3d')
A.contour3D(x,y,z,20)
```

```python
A = plt.axes(projection = '3d')
A.plot_surface(x,y,z)
```

```python
fig = plt.figure()
au = np.array([2,4,6,8,10])
fig,axes = plt.subplots(1,4,figsize= (30,9))
axes[0].set_title('step graph')
axes[0].scatter(au,au**np.pi,color= 'red')
axes[1].set_title('histograhp')
axes[1].hist(au,au,color='blue')
axes[2].set_title('bar grahp')
axes[2].bar(au,au,color='yellow')
axes[3].set_title('line graph')
axes[3].fill_between(au, au**2, au, au**3, color= 'green')
```
