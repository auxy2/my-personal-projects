```python
au = np.array([0,1,2,3,4,5])

fig,axes = plt.subplots(1,6, figsize = (25,7))
axes[0].set_title('Step Graph')
axes[0].step(au, au,lw=5,alpha=1)
axes[1].set_title('Bar Graph')
axes[1].bar(au, au,lw=10,color='r')

# patterns =  ['/','o','*','/','*']
# for bar in axes[1]:
#     axes[1].set_hatch(patterns.pop())
axes[2].set_title('Scatter Graph')
axes[2].scatter(au, au**3,color='orange',lw=3)
axes[3].set_title('Line Graph')
axes[3].plot(au, au**2,'r+-', au, au**3, 'b:o',lw=3 )
axes[4].set_title('Hitgraph')
axes[4].hist(au**2/3, color='y')
axes[5].set_title('Fill_up Graph')
axes[5].fill_between(au ,au**3, au,au**2, color= 'g',lw=3)
axes[0].grid(True)
axes[1].grid(True)
axes[2].grid(True)
axes[3].grid(True)
axes[4].grid(True)
axes[5].grid(True)

```
